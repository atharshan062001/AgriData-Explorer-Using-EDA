```python
query = """
WITH state_rice AS (
  SELECT "State Name", SUM("RICE PRODUCTION (1000 tons)") AS total_rice
  FROM agri_yield_district_wise_data
  GROUP BY "State Name"
  ORDER BY total_rice DESC
  LIMIT 3
)
SELECT a."Year", a."State Name", SUM(a."RICE PRODUCTION (1000 tons)") AS yearly_rice
FROM agri_yield_district_wise_data a
JOIN state_rice s ON a."State Name" = s."State Name"
GROUP BY a."Year", a."State Name"
ORDER BY a."Year", yearly_rice DESC;
"""
result_df = pd.read_sql(query, engine)
print(result_df)
```

```python
query = """
WITH recent_years AS (
  SELECT DISTINCT "Year"
  FROM agri_yield_district_wise_data
  ORDER BY "Year" DESC
  LIMIT 5
),
wheat_yields AS (
  SELECT "Dist Name", "Year", AVG("WHEAT YIELD (Kg per ha)") AS avg_yield
  FROM agri_yield_district_wise_data
  WHERE "Year" IN (SELECT "Year" FROM recent_years)
  GROUP BY "Dist Name", "Year"
),
yield_diff AS (
  SELECT "Dist Name",
  MAX(avg_yield) - MIN(avg_yield) AS yield_increase
  FROM wheat_yields
  GROUP BY "Dist Name"
)
SELECT *
FROM yield_diff
ORDER BY yield_increase DESC
LIMIT 5;

"""
result_df = pd.read_sql(query, engine)
print(result_df)
```

```python
query = """
WITH years AS (
  SELECT DISTINCT "Year" 
  FROM agri_yield_district_wise_data 
  ORDER BY "Year" DESC 
  LIMIT 5
),
oilseed_data AS (
  SELECT "State Name", "Year", SUM("OILSEEDS PRODUCTION (1000 tons)") AS total_prod
  FROM agri_yield_district_wise_data
  WHERE "Year" IN (SELECT "Year" FROM years)
  GROUP BY "State Name", "Year"
),
growth_calc AS (
  SELECT "State Name",
         MAX("Year") AS latest_year,
         MIN("Year") AS oldest_year,
         MAX(total_prod) AS recent,
         MIN(total_prod) AS past
  FROM oilseed_data
  GROUP BY "State Name"
),
final_growth AS (
  SELECT "State Name",
         ((recent - past) * 100.0 / past) AS growth_rate
  FROM growth_calc
  WHERE past > 0
)
SELECT * 
FROM final_growth
ORDER BY growth_rate DESC
LIMIT 5;

"""
result_df = pd.read_sql(query, engine)
print(result_df)
```

```python
query = """
SELECT "Dist Name",
       corr("RICE AREA (1000 ha)", "RICE PRODUCTION (1000 tons)") AS rice_corr,
       corr("WHEAT AREA (1000 ha)", "WHEAT PRODUCTION (1000 tons)") AS wheat_corr,
       corr("MAIZE AREA (1000 ha)", "MAIZE PRODUCTION (1000 tons)") AS maize_corr
FROM agri_yield_district_wise_data
GROUP BY "Dist Name";

"""
result_df = pd.read_sql(query, engine)
print(result_df)
```

```python
query = """
WITH top_cotton_states AS (
  SELECT "State Name", SUM("COTTON PRODUCTION (1000 tons)") AS total_cotton
  FROM agri_yield_district_wise_data
  GROUP BY "State Name"
  ORDER BY total_cotton DESC
  LIMIT 5
)
SELECT "Year", a."State Name", SUM("COTTON PRODUCTION (1000 tons)") AS yearly_production
FROM agri_yield_district_wise_data a
JOIN top_cotton_states t ON a."State Name" = t."State Name"
GROUP BY "Year", a."State Name"
ORDER BY "Year", a."State Name";

"""
result_df = pd.read_sql(query, engine)
print(result_df)
```

```python
query = """
SELECT "Dist Name", SUM("GROUNDNUT PRODUCTION (1000 tons)") AS total_production
FROM agri_yield_district_wise_data
WHERE "Year" = 2017
GROUP BY "Dist Name"
ORDER BY total_production DESC
LIMIT 10;

"""
result_df = pd.read_sql(query, engine)
print(result_df)
```

```python
query = """
WITH top_states AS (
  SELECT "State Name", SUM("RICE PRODUCTION (1000 tons)" + "WHEAT PRODUCTION (1000 tons)") AS total
  FROM agri_yield_district_wise_data
  GROUP BY "State Name"
  ORDER BY total DESC
  LIMIT 5
)
SELECT "Year", "State Name",
       SUM("RICE PRODUCTION (1000 tons)") AS rice_prod,
       SUM("WHEAT PRODUCTION (1000 tons)") AS wheat_prod
FROM agri_yield_district_wise_data
WHERE "State Name" IN (SELECT "State Name" FROM top_states)
GROUP BY "Year", "State Name"
ORDER BY "Year", "State Name";

"""
result_df = pd.read_sql(query, engine)
print(result_df)
```
